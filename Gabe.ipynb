```python
import pandas as pd
```

```python
# read in people dataset
people = pd.read_csv('./data/FL/final_people.csv')
people.head()
```

```python
# check people data types
people.dtypes
```

```python
# convert all datetime columns
people['birth_date'] = pd.to_datetime(people['birth_date'])
people['first_jailing_date'] = pd.to_datetime(people['first_jailing_date'])
people['last_jailing_date'] = pd.to_datetime(people['last_jailing_date'])
people['last_jailing_release'] = pd.to_datetime(people['last_jailing_release'])
people['first_incarceration_date'] = pd.to_datetime(people['first_incarceration_date'])
people['last_incarceration_date'] = pd.to_datetime(people['last_incarceration_date'])
people['last_incarceration_release'] = pd.to_datetime(people['last_incarceration_release'])
```

```python
# final dtype/null check
people.info()
# looks like we only have null issues for folks who were never jailed and/or imprisoned
# Q for group - what to do with these?
# A: combine jailing/incarceration - if no incarceration, use jailing; drop people with neither
```

```python
people['first_incarceration_date'] = people['first_incarceration_date'].combine_first(people['first_jailing_date'])
people['last_incarceration_date'] = people['last_incarceration_date'].combine_first(people['last_jailing_date'])
people['last_incarceration_release'] = people['last_incarceration_release'].combine_first(people['last_jailing_release'])
people['num_incarcerations'] += people['num_jailings']
```

```python
people.info()
```

```python
# now we can dropna and drop the jailing columns
people = people.drop(columns = ['num_jailings', 'first_jailing_date', 'last_jailing_date', 'last_jailing_release']).dropna()
```

```python
people.info()
```

```python
# read in charges dataset
charges = pd.read_csv('./data/FL/final_charges.csv')
charges.head()
```

```python
charges.dtypes
```

```python
# convert datetimes
charges['offense_date'] = pd.to_datetime(charges['offense_date'])
charges['arrest_date'] = pd.to_datetime(charges['arrest_date'])
```

```python
# final dtypes/null check
charges.info()
# looks like we have some null charges, those can probably be dropped safely
# but there are a lot of null arrest dates
# Q for group - same as above
```

```python
# for the charges - what's the difference between offense date and arrest date?
charges[(charges['offense_date'] != charges['arrest_date']) & (charges['arrest_date'].notnull())]
# looks like we have a lot of differences - arrest date is an arbitrary amount of time after offense
# we'll keep both for now
```

```python
# so how do we combine these into one useful dataset?
# proposal for charges:
# keep earliest offense, charge degree, charge, and arrest date as columns
# keep latest offense, charge degree, charge, and arrest date as columns
# pivot charge_degree into count columns per degree
# keep a concatenated version of all charge descriptions for NLP
# also create a column of the mean time between offenses and the mean time between arrests
# ending up with one row per person to be joined with main people dataset
```

```python
people['first_offense_date'] = \
    people.join(charges.sort_values(by = ['person_id', 'offense_date']).groupby('person_id')['offense_date'].first(),
    on = 'person_id',
    how = 'left')['offense_date']
```

```python
people['first_charge_degree'] = \
    people.join(charges.sort_values(by = ['person_id', 'offense_date']).groupby('person_id')['charge_degree'].first(),
    on = 'person_id',
    how = 'left')['charge_degree']
```

```python
people['first_charge'] = \
    people.join(charges.sort_values(by = ['person_id', 'offense_date']).groupby('person_id')['charge'].first(),
    on = 'person_id',
    how = 'left')['charge']
```

```python
people['first_arrest_date'] = \
    people.join(charges.sort_values(by = ['person_id', 'offense_date']).groupby('person_id')['arrest_date'].first(),
    on = 'person_id',
    how = 'left')['arrest_date']
```

```python
people['last_offense_date'] = \
    people.join(charges.sort_values(by = ['person_id', 'offense_date']).groupby('person_id')['offense_date'].last(),
    on = 'person_id',
    how = 'left')['offense_date']
```

```python
people['last_charge_degree'] = \
    people.join(charges.sort_values(by = ['person_id', 'offense_date']).groupby('person_id')['charge_degree'].last(),
    on = 'person_id',
    how = 'left')['charge_degree']
```

```python
people['last_charge'] = \
    people.join(charges.sort_values(by = ['person_id', 'offense_date']).groupby('person_id')['charge'].last(),
    on = 'person_id',
    how = 'left')['charge']
```

```python
people['last_arrest_date'] = \
    people.join(charges.sort_values(by = ['person_id', 'offense_date']).groupby('person_id')['arrest_date'].last(),
    on = 'person_id',
    how = 'left')['arrest_date']
```

```python
# weird that we have to select a column out of this
charge_pivot = pd.pivot_table(charges, index = 'person_id', columns = 'charge_degree', aggfunc = 'count')['arrest_date']
```

```python
# let's fix the names
charge_pivot.columns = ['charge_degree_' + col.replace('(', '').replace(')', '') for col in charge_pivot.columns]
```

```python
people = people.join(charge_pivot, on = 'person_id', how = 'left')
```

```python
people.info()
```

```python
# https://stackoverflow.com/questions/27298178/concatenate-strings-from-several-rows-using-pandas-groupby
agg_charges = charges[charges['charge'].notnull()].groupby('person_id').agg({'charge': '|'.join})
```

```python
agg_charges
```

```python
people['agg_charges'] = \
    people.join(agg_charges, on = 'person_id', how = 'left')['charge']
```

```python
people.head()
```

```python
# https://stackoverflow.com/questions/45241221/python-pandas-calculate-average-days-between-dates
charges['previous_offense'] = \
    charges.sort_values(by = ['person_id', 'offense_date']).groupby(['person_id'])['offense_date'].shift(1)
charges['days_between_offenses'] = \
    (charges['offense_date'] - charges['previous_offense']).apply(lambda x: x.days)
```

```python
charges['previous_arrest'] = \
    charges.sort_values(by = ['person_id', 'arrest_date']).groupby(['person_id'])['arrest_date'].shift(1)
charges['days_between_arrests'] = \
    (charges['arrest_date'] - charges['previous_arrest']).apply(lambda x: x.days)
```

```python
charges
```

```python
# need to remove zeros - don't want them affecting the averages
days_between_offenses = charges[charges['days_between_offenses'] > 0].groupby('person_id')['days_between_offenses'].mean()
days_between_arrests = charges[charges['days_between_arrests'] > 0].groupby('person_id')['days_between_arrests'].mean()
```

```python
people['avg_days_between_offenses'] = \
    people.join(days_between_offenses, on = 'person_id', how = 'left')['days_between_offenses']
people['avg_days_between_arrests'] = \
    people.join(days_between_arrests, on = 'person_id', how = 'left')['days_between_arrests']
```
